Handle Missing Values

```python
df['DriveTrain'].mode()
```

```python
df['AirBags'].mode()
```

```python
df['Manufacturer'].mode()[0]
```

```python
for i in df.columns:
    if df[i].dtype=='object': # seggregate the cat column
        m=df[i].mode()[0] # Storing the most frequent in var m
        df[i]=df[i].fillna(m)  # finding missingvalues , replacing it with modal value and updating the df
    else:
        me=df[i].mean()
        df[i]=df[i].fillna(me)  # fill missing values with mean of the respective column
```
